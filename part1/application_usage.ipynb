# Histogramme de trois premiers Users avec leur utilisation d'écrans

```python
nb_users_df_usage = df_usage['user_id'].nunique()
first_users_df_usage = df_usage['user_id'].unique()[:3]

for uid in first_users_df_usage:
    df_user = df_usage[df_usage['user_id'] == uid]

    nb_jours = df_user['date'].nunique()
    total_screen_time = df_user['screen_time_min'].sum()
    moyenne_par_jour = total_screen_time / nb_jours if nb_jours > 0 else 0

    min_time = df_user['screen_time_min'].min()
    max_time = df_user['screen_time_min'].max()
    mean_session = df_user['screen_time_min'].mean()

    # Prepare graph
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))

    # Histogram on the left
    sns.histplot(df_user['screen_time_min'], bins=20, kde=True, ax=axes[0])
    axes[0].set_title(f'User {uid}\nAvg/day: {moyenne_par_jour:.1f} min over {nb_jours} days')
    axes[0].set_xlabel('Minutes per session')
    axes[0].set_ylabel('Number of sessions')

    # Barplot of top apps on the right
    top_apps = df_user.groupby('app_name')['screen_time_min'].sum().sort_values(ascending=False).head(10)
    sns.barplot(x=top_apps.values, y=top_apps.index, ax=axes[1])
    axes[1].set_title(f'Top 10 apps - User {uid}')
    axes[1].set_xlabel('Total time (min)')
    axes[1].set_ylabel('Application')

    plt.tight_layout()
    plt.show()
```

# Les applications les plus utilisées par User

```python
df_prod = df_usage[df_usage['is_productive'] == True]
df_nonprod = df_usage[df_usage['is_productive'] == False]

# Calcul du temps total par application pour chaque catégorie
total_time_prod = df_prod.groupby('app_name')['screen_time_min'].sum().sort_values(ascending=False)
total_time_nonprod = df_nonprod.groupby('app_name')['screen_time_min'].sum().sort_values(ascending=False)

# Productives
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.pie(total_time_prod.values, labels=total_time_prod.index, autopct='%1.1f%%', startangle=140)
plt.title("Productive apps")
plt.axis('equal')

# Non productives
plt.subplot(1, 2, 2)
plt.pie(total_time_nonprod.values, labels=total_time_nonprod.index, autopct='%1.1f%%', startangle=140)
plt.title("Non-productive apps")
plt.axis('equal')

plt.tight_layout()
plt.show()
```

# Temps passé de façon productive ou non

```python
# Calcul du temps total pour chaque catégorie productive/non productive
total_productive = df_usage[df_usage['is_productive'] == True]['screen_time_min'].sum()
total_nonproductive = df_usage[df_usage['is_productive'] == False]['screen_time_min'].sum()
total = total_productive + total_nonproductive

pourc_productive = 100 * total_productive / total if total > 0 else 0
pourc_nonproductive = 100 * total_nonproductive / total if total > 0 else 0

df_usage['date'] = pd.to_datetime(df_usage['date'])

# Jour minimum et maximum
date_min = df_usage['date'].min()
date_max = df_usage['date'].max()
nb_jours_total = (date_max - date_min).days + 1

print(f"Data ranges from {date_min.date()} to {date_max.date()}")
print(f"Total number of days covered            : {nb_jours_total}")
print(f"Number of users included                : {nb_users_df_usage}")
print(f"Total productive time                   : {total_productive:.2f} min ({pourc_productive:.1f}%)")
print(f"Total non-productive time               : {total_nonproductive:.2f} min ({pourc_nonproductive:.1f}%)")
```

# Liste des applications considérées "dangereuses"

```python
# https://www.bark.us/blog/dangerous-apps-kids/?srsltid=AfmBOooDW775NTcJBS84NwsyYsmAY9SN7oQAEFD-QunIAwSswaDTe-IY
bark_dangerous_apps = ["Snapchat", "Instagram", "Discord", "Yik Yak", "TikTok", "Twitter", "Omegle", "Kik", "Telegram", "Hoop", "Vault apps", "Roblox"]

# https://safelagoon.com/en/blog/insights/30-most-dangerous-apps-for-kids-in-2024/
safelagoon_dangerous_apps = ["TikTok", "Snapchat", "Ask.fm", "Kik", "Tinder", "Instagram", "Houseparty", "Holla", "Blendr", "Zoomerang", "Parlor", "Hoop", "Voxer", "Tumblr", "Bumble", "VSCO", "BIGO LIVE", "Confide", "Omegle", "IMVU", "WeChat", "Telegram", "Periscope", "Roblox", "Twitter", "Yarn", "Chat Master", "Among Us", "Tellonym", "Vault apps"]

# https://parental-control.net/en/blog/article/top-7-most-dangerous-apps-for-kids-in-2025-what-parents-should-know
parental_control_dangerous_app = ["TikTok", "Discord", "Snapchat", "Roblox", "Instagram", "Omegle", "Telegram"]
```

```python
# Combinaison et comptage manuel des occurrences
all_apps = bark_dangerous_apps + safelagoon_dangerous_apps + parental_control_dangerous_app
app_counts = {}

for app in all_apps:
    if app in app_counts:
        app_counts[app] += 1
    else:
        app_counts[app] = 1

# Tri des applications par fréquence décroissante
sorted_app_counts = dict(sorted(app_counts.items(), key=lambda x: x[1], reverse=True))

# Préparation des données pour le graphique
apps = list(sorted_app_counts.keys())
frequencies = list(sorted_app_counts.values())

# Couleurs : top 3 en rouge, orange et or
colors = ['red' if i == 0 else 'orange' if i == 1 else 'gold' if i == 2 else 'skyblue' for i in range(len(apps))]

# Création du graphique
plt.figure(figsize=(14, 10))
bars = plt.barh(apps, frequencies, color=colors)

# Ajouter les étiquettes de données
for bar in bars:
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}', va='center', fontsize=10)

# Ajouts esthétiques
plt.xlabel("Occurrences Across Sources", fontsize=12)
plt.title("Top Dangerous Apps for Teens (2024-2025)", fontsize=16, weight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
dangerous_apps = set(bark_dangerous_apps + safelagoon_dangerous_apps + parental_control_dangerous_app)

# Liste des apps présentes dans df_usage
used_apps = set(df_usage.groupby('app_name').groups.keys())

# Intersection : apps à la fois dangereuses et utilisées
dangerous_used_apps = sorted(dangerous_apps.intersection(used_apps))

# Affichage
print("Applications considered as dangerous in df_usage :")
print(dangerous_used_apps)
```
